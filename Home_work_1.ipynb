```python
# Question 4

count = 0

for i in  range(num_columns):
  if fuel_data.isnull().sum().iloc[i]>0:
    count+=1

print(count) # Answer = 4
fuel_data.isnull().sum()
```

```python
# Question 5
round(fuel_data[fuel_data["origin"] == "Asia"]["fuel_efficiency_mpg"].max(), 2) # Answer = 23.75
```

```python
# Question 6
median = fuel_data["horsepower"].median() # Median 149
modd = fuel_data["horsepower"].mode() # Mode = 152
fuel_data["horsepower"] = fuel_data["horsepower"].fillna(value=modd)
sec_median = fuel_data["horsepower"].median() # Second median 149
```

```python
# Question 7

all_asian_cars = fuel_data[fuel_data["origin"] == "Asia"]
asian_weight_model = all_asian_cars[["vehicle_weight", "model_year"]]
first_sev = asian_weight_model[:7]
X = first_sev.to_numpy()
X_T = X.T
XTX= np.matmul(X_T,X)
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = XTX_inv @ X_T @ y
result = round(w.sum(),2)

print(result) # Answer = 0.51
```
